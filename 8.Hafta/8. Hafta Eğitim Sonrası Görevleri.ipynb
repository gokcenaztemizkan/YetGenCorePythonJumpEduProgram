# Problem 1

Şimdi beraber kullanıcılardan aldığımız verileri işleyerek, onlara hem çıktı verecek, hem de aldığımız ve hesapladığımız verileri ileride kullanabilmek adına kaydedeceğiz.


 **Bu bir veri giriş platformu olacak**


 * kullanıcıya kaç giriş yapacağını soracağız
* kullanıcıdan isim-soyisim, yaş, boy, kilo bilgilerini alarak VKİ hesaplayacağız
* ardından bunları sınıflara göre kaydedeceğiz bir dosyaya kaydedeceğiz


Sınıflar ise şu şekilde: 
* 18, 5 kg/m.'nin altında olanlar: Zayıf
* 18.5 – 24, 9 kg/m. arasında olanlar: Normal kilolu
* 25 – 29, 9 kg/m. arasında olanlar: Fazla kilolu
* 30 – 39, 9 kg/m. arasında olanlar: Obez
* 40 kg/m.'nin üzerinde olanlar: İleri derecede obez (morbid obez), olarak görülür.

```python
# 1.Problemin Çözümünü Buraya Yazınız
giris_sayisi = int(input("Kaç giriş yapacaksınız? "))

kullanici_verileri = []

for i in range(giris_sayisi):
    print(f"Giriş {i+1}")
    isim_soyisim = input("İsim ve Soyisim: ")
    yas = int(input("Yaş: "))
    boy = float(input("Boy (metre cinsinden): "))
    kilo = float(input("Kilo (kilogram cinsinden): "))
    
    vki = kilo / (boy**2)
    
    sinif = ""
    if vki < 18.5:
        sinif = "Zayıf"
    elif 18.5 <= vki <= 24.9:
        sinif = "Normal kilolu"
    elif 25 <= vki <= 29.9:
        sinif = "Fazla kilolu"
    elif 30 <= vki <= 39.9:
        sinif = "Obez"
    else:
        sinif = "İleri derecede obez (morbid obez)"
    
    kullanici_veri = {"İsim ve Soyisim": isim_soyisim, "Yaş": yas, "Boy (m)": boy, "Kilo (kg)": kilo, "VKİ": vki, "Sınıf": sinif}
    kullanici_verileri.append(kullanici_veri)

print("Kaydedilen Kullanıcı Verileri:")
for i, kullanici in enumerate(kullanici_verileri):
    print(f"Giriş {i+1}:")
    print(f"İsim ve Soyisim: {kullanici['İsim ve Soyisim']}")
    print(f"Yaş: {kullanici['Yaş']}")
    print(f"Boy (m): {kullanici['Boy (m)']}")
    print(f"Kilo (kg): {kullanici['Kilo (kg)']}")
    print(f"VKİ: {kullanici['VKİ']}")
    print(f"Sınıf: {kullanici['Sınıf']}")
    print("-------")
    
with open("kullanici_verileri.txt", "w") as dosya:
    for kullanici in kullanici_verileri:
        dosya.write(f"İsim ve Soyisim: {kullanici['İsim ve Soyisim']}\n")
        dosya.write(f"Yaş: {kullanici['Yaş']}\n")
        dosya.write(f"Boy (m): {kullanici['Boy (m)']}\n")
        dosya.write(f"Kilo (kg): {kullanici['Kilo (kg)']}\n")
        dosya.write(f"VKİ: {kullanici['VKİ']}\n")
        dosya.write(f"Sınıf: {kullanici['Sınıf']}\n")
        dosya.write("-------\n")
```

# Problem 2
**YetBar - Smoothie Bank**
---
YetBar spor salonunun girişinde bulunan, insansız bir sağlıklı içecek otomatıdır. Kullanıcı çeşitli girdiler smoothie içeceğine koyacağı ürünleri belirleyebilir, fiyat hesaplaması yaptırabilir ya da sadece menüyü görüntüleyebilir.

**Projenin Amacı:**

*   Birbirini etkileyen ve bir kod içerisinde birden fazla defa başvurulacak fonksiyonların oluşturulup, "fonksiyonel" proglamlamak.
*   Büyük programları yazarken, yazılımcı olarak yapabileceğimiz hatalardan kaçınmak için kontrol yöntemleri geliştirmek.
*   Kullanıcı inputlarını ve yaptığımız hesapları, raporlamak/loglamak amacıyla kaydetmek.

**Programdan Beklenen Çıktılar**
* Program kullanıcıya temelde 2 farklı işlem yapma imkanı sağlar.
> 1. Menüyü görüntüleme
> 3. Satın alma
* Satın alınan her içecek tarih, fiyat bilgisi ile kaydedilmeli ve her zaman ulaşılabilir olmalıdır.




```python
import datetime

urunler = {
    "1": {"isim": "Mango Smoothie", "fiyat": 10},
    "2": {"isim": "Çilek Smoothie", "fiyat": 12},
    "3": {"isim": "Ananas Smoothie", "fiyat": 8},
    "4": {"isim": "Karışık Meyveli Smoothie", "fiyat": 15},
    "5": {"isim": "Yeşil Detox Smoothie", "fiyat": 18},
}

satin_alinanlar = []


def menu_goruntule():
    print("***** YetBar Smoothie Bank Menüsü *****")
    for urun_kodu, urun_bilgileri in urunler.items():
        print(f"{urun_kodu}. Ürün: {urun_bilgileri['isim']} - Fiyat: {urun_bilgileri['fiyat']} TL")


def satin_alma():
    menu_goruntule()
    secim = input("Lütfen satın almak istediğiniz ürünün kodunu girin (Çıkış için 'q'): ")
    while secim != "q":
        if secim in urunler:
            urun_kodu = secim
            urun = urunler[urun_kodu]
            satin_alinanlar.append(
                {"tarih": datetime.datetime.now(), "urun": urun["isim"], "fiyat": urun["fiyat"]}
            )
            print(f"{urun['isim']} seçildi. {urun['fiyat']} TL ödenecek.")
        else:
            print("Geçersiz bir ürün kodu girdiniz.")
        secim = input("Lütfen satın almak istediğiniz ürünün kodunu girin (Çıkış için 'q'): ")


def satin_alinanlari_kaydet():
    with open("satin_alinanlar.txt", "w") as dosya:
        dosya.write("***** Satın Alınan Ürünler *****\n")
        for satin_alinan in satin_alinanlar:
            dosya.write(f"Tarih: {satin_alinan['tarih']}\n")
            dosya.write(f"Ürün: {satin_alinan['urun']}\n")
            dosya.write(f"Fiyat: {satin_alinan['fiyat']} TL\n")
            dosya.write("-------\n")
    print("Satın alınan ürünler kaydedildi!")


while True:
    print("***** YetBar Smoothie Bank *****")
    print("1. Menüyü görüntüle")
    print("2. Satın alma")
    print("3. Çıkış")
    secim = input("Lütfen bir seçenek girin: ")
    if secim == "1":
        menu_goruntule()
    elif secim == "2":
        satin_alma()
    elif secim == "3":
        satin_alinanlari_kaydet()
        print("Çıkıyor...")
        break
    else:
        print("Geçersiz bir seçenek girdiniz. Lütfen tekrar deneyin.")
```
